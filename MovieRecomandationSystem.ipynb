```python
titles=list(data.get('title'))
print(type(titles))
print(titles[:5])
```

```python
freq_titles={}
for title in titles:
  length=len(title)
  if freq_titles.get(length) is None:
      freq_titles[length]=1
  else:
      freq_titles[length]
```

```python
print(freq_titles)
```

```python
x=np.array(list(freq_titles.keys()))
y=np.array(list(freq_titles.values()))
plt.scatter(x,y)
plt.xlabel("length of movie title")
plt.ylabel("freq count")
plt.title("movie data visuliation problem")
plt.show()
```
