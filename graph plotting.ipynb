```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
#adjust the size of any plot
plt.figure(figsize=(2,2))
plt.plot(x,y1,color='red',label='apple',marker='o')
plt.plot(x,y2,color='grey',label='kiwi',linestyle="dashed",marker='*')
plt.legend()
plt.title("Harsh ki bikri")
plt.xlabel("time")
plt.ylabel("price")

plt.show()
```

```python
prices=np.array([1,2,3,4])**3
print(prices)
plt.plot(prices)
```

```python
#BAR GRAPH
x=np.array([0,2,4])
plt.style.use("dark_background")
plt.bar(x,[10,20,15],width=0.4,label="current year",tick_label=["gold","platinum","silver"])
plt.bar(x+0.5,[20,10,12],width=0.4,label="next year",color="orange")
plt.xlabel("metal")
plt.ylabel("price")
plt.title("Metal price comparison")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,40)
plt.show()
```

```python
#pie chart
plt.style.use("seaborn")
subjects="maths","chem","physics","english"
weightage=[20,10,21,34]
plt.pie(weightage,labels=subjects,explode=(0,0,0,0.2),shadow=True)
plt.show()
```
